# matplotlib 연습

```python
import numpy as np
import matplotlib.pyplot as plt
```

## plt.show()

```python
data = np.arange(1, 100)
plt.plot(data)
plt.show()  # print같은거라고 생각하면 됨. print() 안찍어도 출력이 되긴한다
```

```python
plt.plot(data)  # plt.show()안찍으면 [<matplolib.lines.Line2D at ~~~~>]이 나오는것 확인 가능
```

## 그래프 두개 그리기

```python
data = np.arange(1, 51)
plt.plot(data)
data2 = np.arange(51, 101)
plt.plot(data2)  # plt.plot을 두번찍었더니 하나의 창 안에 두개의 그래프가 그려짐
plt.show()
```

## plt.figure

```python
# plt.figure는 새로운 창을 만드는 것. 앞에 봐서 알겠지만 만들지 않는다고 해서 그래프가 안그려지는 것은 아님.
fig = plt.figure()
# 하지만 만들고나서 할당을하면 다음과 같은 세부 작업이 가능함.
fig.set_size_inches(10, 5)

data = np.arange(100, 201)
plt.plot(data)

data2 = np.arange(200, 301)
fig2 = plt.figure()  # 이건 새로운 그래프(창)를 생성함.
plt.plot(data2)

plt.show()
```

## plt.subplot (하나의 figure안에서 창 나누기)

```python
data = np.arange(100, 201)
plt.subplot(2, 1, 1)  # 하나의 창을 행2개 열1개로 쪼갠 뒤 1번째 칸에 plot을 그릴거다.
plt.plot(data)

data2 = np.arange(200, 301)
plt.subplot(2, 1, 2)
plt.plot(data2)

plt.show()
```

```python
data = np.arange(100, 201)
plt.subplot(211)  # 콤마를 제거해도 똑같이 작동함
plt.plot(data)

data2 = np.arange(200, 301)
plt.subplot(212)
plt.plot(data2)

plt.show()
```

```python
data = np.arange(100, 201)
plt.subplot(1, 3, 1)  # 하나의 창을 행1개 열3개로 쪼갠 후 1번째 칸에 다음에 나올 plt.plot()들을 그릴거다.
plt.plot(data)

data2 = np.arange(200, 301)
plt.subplot(1, 3, 2)
plt.plot(data2)

data3 = np.arange(300, 401)
plt.subplot(1, 3, 3)
plt.plot(data3)

plt.show()
```

## plt.subplots (s가 하나 더 붙는다)

```python
data = np.arange(1, 51)

fig, axes = plt.subplots(2, 3)
# fig, axes는 그냥 리턴이 두개인거라 꼭 줘야함. 여기서 fig는 창6개를 모두 포함한 하나의 통합창, axes는 창 하나씩을 의미한다
# plt.subplots를 쓰면 plt.figure()를 자동으로 해준다고 생각하면 됨.
fig.set_size_inches(15, 10)

axes[0, 0].plot(data)
axes[0, 0].plot(data % 10)  # 하나에 여러개 그리고 싶으면 이렇게
axes[0, 1].plot(data * data)
axes[0, 2].plot(data ** 3)
axes[1, 0].plot(data % 10)
axes[1, 1].plot(-data)
axes[1, 2].plot(data // 20)

fig.tight_layout()  # 이 코드를 넣으면 창끼리 겹치는걸 방지해줌. 궁금하면 fig.set_size_inches(15, 10) 여기 사이즈 줄이고 빼볼 것.
plt.show()
```

## Label 설정

```python
plt.plot([1, 2, 3], [3, 6, 9])
plt.plot([1, 2, 3], [2, 4, 9])

# 타이틀 & font 설정
plt.title('Label', fontsize=20)  # 제목으로 한글주면 오류나는데 리눅스라서 그런가..

# X축 & Y축 Label 설정
plt.xlabel('X', fontsize=20)
plt.ylabel('Y', fontsize=20)

plt.show()
```

## 눈금(Tick), 범례(Legend) 설정

```python
plt.plot(np.arange(10), np.arange(10)*2)
plt.plot(np.arange(10), np.arange(10)**2)
plt.plot(np.arange(10), np.log(np.arange(1, 11)))

plt.title('Tick & Legend', fontsize=20)

plt.xlabel('X', fontsize=20)
plt.ylabel('Y', fontsize=20)

# Tick 설정 (눈금)
plt.xticks(rotation=90)
plt.yticks(rotation=30)

# legend 설정 (범례)
plt.legend(['10 * 2', '10 ** 2', 'log'], fontsize=15)

plt.show()
```

## Limit 설정

```python
fig = plt.figure()
fig.set_size_inches(10, 10)

plt.plot(np.arange(10), np.arange(10)*2)
plt.plot(np.arange(10), np.arange(10)**2)
plt.plot(np.arange(10), np.log(np.arange(1, 11)))

plt.title('Limit', fontsize=20)

plt.xlabel('X', fontsize=20)
plt.ylabel('Y', fontsize=20)

# Tick 설정 (눈금)
plt.xticks(rotation=90)
plt.yticks(rotation=30)

# legend 설정 (범례)
plt.legend(['10 * 2', '10 ** 2', 'log'], fontsize=15)

# x, y limit 설정
plt.xlim(0, 5)
plt.ylim(0, 5)

plt.show()
```
